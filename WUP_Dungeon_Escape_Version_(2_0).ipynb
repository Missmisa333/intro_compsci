```python
#@title
import time
from IPython.display import clear_output 

#The possible choices the user can choose from are a,A,b,B,c, and C
lower = ['a','b','c']
upper = ['A','B','C']

#Give a dummy value to choice since it doesn't get defined until the user choses an action after the choose the mode
choice = ''

#The correct answer choice to escape & the location in the list corresponds to the mode
success_choice_1 = [2,2,1]


def data_validation(user_mode_choice,lower,upper):
  while user_mode_choice not in lower and user_mode_choice not in upper:
    print("Please select from A,B, or C")
    user_mode_choice = input()

  if user_mode_choice in lower:
    user_mode_choice = user_mode_choice.upper()
  return user_mode_choice


def create_choice_number(user_mode_choice):
  if user_mode_choice == 'A':
    number = 0
  elif user_mode_choice == 'B':
    number = 1
  else:
    number = 2
  return number


def give_options(action_1, mode_number, upper):
  for option, upper in zip(action_1[mode_number],upper):
    print(upper, option[0])



#This is the template for the synopsys
starting_action = 'You wake up in a locked, {}, {}. \nThe only thing by you is a {}'


#the 1st item in the following lists corresponds to the elevator mode (mode_number = 0). 
#the 2nd item in the following lists corresponds to the Dungeon Cell mode (mode_number = 1). 
#the 3rd item in the following lists corresponds to the Haunted House mode (mode_number = 2). 
#For example, the elevator mode will get 'actually quite nice', 'elevator', and 'carboard box' from the lists below.
location_description = ['actually quite nice','dark','haunted']
location = ['elevator', 'dungeon cell', 'mansion']
thing = ['carboard box', 'ring of keys', 'an old map']

#The following are the actions that can be taken and the responses to them
#It is labeled action_1 so if I want to add to it later, it is easy to create, for example, action_2

action_1 = [
            [['Look at what is inside the cardboard box.', 'There is nothing in the box'],
            ['Climb above on the top of the elevator to see if you can lift the ceilling.', 'You lift the ceiling but it is not the movies. You are too scared to get up there.'],
            ['Call the help button on the elevator.','The rescue team answered came and rescued you!!!']],
            [['Try to see if one of the keys will open the lock.', 'An alarm goes off and a guard comes to take the keys away from you.'],
             ['Shake the dungeon door to see if it will budge.', 'A trap is activated and you now have two sets of locked doors instead of just one.'],
             ['Since it is dark, take a nap.','A hero comes and rescues you!!!']],
            [['Investigate the map to see if there are any useful clues on it.', 'The map is too old to be any use.'],
             ['Call out to the ghosts in the mansion to see if they will help you.','The ghosts show you to a hidden door so you can get out!'],
             ['Start checking every door & window in the house to see if you can find something useful.','You find nothing.']]            
            ]
#Description for the structure above: 
'''
action = [
  [#mode 0
    ['choice A', 'response to A'], 
    ['choice B', 'response to B'],
    ['choice C', 'response to C']],
  [#mode 1
    ['choice A', 'response to A'], 
    ['choice B', 'response to B'],
    ['choice C', 'response to C']],
  [#mode 2
    ['choice A', 'response to A'], 
    ['choice B', 'response to B'],
    ['choice C', 'response to C']]]


so the syntax is action[mode number][choice number][0 for action or 1 for response] 
choice number is 0 for A, 1 for B, and 2 for C

so action[0][2][1] will get you the response for choice C in the elevator mode
'''

#Ask the user to choose a mode
print('Select a mode: ')
print('A: Elevator')
print('B: Dungeon Cell')
print('C: Haunted House')

user_mode_choice = input()
user_mode_choice = data_validation(user_mode_choice,lower,upper)
#Convert the user's choice into a mode_number for easy access
mode_number = create_choice_number(user_mode_choice)
clear_output(wait=True)
#Load the action that would get you successfully rescued or escaped
success = action_1[mode_number] [success_choice_1[mode_number]] [0]


#Fill in the starting_action with the location description, location, and item around you based off of the mode the user chose
starting_action = starting_action.format(location_description[mode_number], location[mode_number],thing[mode_number] )
print(starting_action)

time.sleep(2)

#While the user's choice is not the choice that successfully rescues or escapes you
while choice != success:
  print('___________________________________')
  print("What would you like to do?")
  print(" ")
  #Give the A,B,C choices to the user. This function will print the A,B,C options for that mode 
  give_options(action_1, mode_number, upper)
  #Have the user choose A,B, or C
  user_action_1 = input()
  user_action_1 = data_validation(user_action_1,lower,upper)

  #Store a choice number based off whether they choose A,B or C. A is 0, B is 1 and C is 2
  user_action_1_choice = create_choice_number(user_action_1)
  #Load the user's choice & the reaciton that is tied to that choice into the variables choice & reaction 
  choice = action_1[mode_number][user_action_1_choice][0]
  reaction = action_1[mode_number][user_action_1_choice][1]
  
  time.sleep(1)
  clear_output(wait=True)
  #Give the user the response
  print('\n',reaction)
  time.sleep(1)
  #If the user's choice is the one that successfully rescues them then they can leave the while loop 
  #If not, we will remove that choice from the list of choices that the user can choose from. And have the user choose again.
  action_1[mode_number].remove([choice,reaction])
```
