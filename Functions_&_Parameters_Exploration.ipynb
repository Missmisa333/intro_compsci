##Ordering at a Cafe

```python
#Cell 2

def get_price(order):
  
  if order == "A":
    order = "Cheesecake"
    price = 5.52
    price_with_tax = price * 1.07
    price_with_tax = round(price_with_tax, 2)

  elif order == "B":
    order = "Lavender White Chocolate Mocha"
    price = 3.52
    price_with_tax = price * 1.07
    price_with_tax = round(price_with_tax, 2)

  else:
    order = "None"
    price = "None"
    price_with_tax = "None"

  return order, price, price_with_tax



def make_sure_its_on_the_menu(order):

  while order!= "A" and order!= "B":
    print("That is not a valid order.")
    print("Please select betwen A: Cheesecake or B: Lavender White Chocolate Mocha")
    order = input()




print("Welcome to our cafe!!")
print("Please between our two fan favorites! ")
print("A: Cheesecake" )
print("B: Lavendar White Chocolate Mocha")

order = input()

validated_order = make_sure_its_on_the_menu(order)
order, price, price_with_tax = get_price(validated_order)

print("So your", validated_order, "costs", price, "dollars and with tax that will be $", price_with_tax)
```
